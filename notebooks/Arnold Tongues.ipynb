### Figure 3: Arnold tongues in the $\tau$-$a$ plane

Data files were created using a C script that simulated the infectious timeseries and computed the average number of peaks per year in the steady state. 

```python
arnold1 = '../simulations/cluster/arnoldtongues/arnold_a=0-30_tau=0-70_mmax=084.dat'
arnold2 = '../simulations/cluster/arnoldtongues/arnold_a=0-30_tau=0-70_mmax=086.dat'

res_a1 = pd.read_csv(arnold1,sep=',')
res_a2 = pd.read_csv(arnold2,sep=',')
```

#### For $m_{max} = 0.84$

```python

cm = 1/2.54
fig,axs = plt.subplots(2,1,figsize=(8.6*cm,8.6*cm),height_ratios=[1,5])
axs[0].axis('off')

show, M, tauline, sline = plotting.plot_arnold(axs[1],res_a1,cmap=cmap,param='a',tol=0.015,windingnumbers=windingnumbers)

cbar = plt.colorbar(show,orientation='horizontal',ax=axs[0],location='top',aspect=15,fraction=0.8,pad=0.15)
cbar.set_ticks(np.arange(1.5,len(windingnumbers)+2))
cbar.ax.set_xticklabels(windingnumberlabels);

axs[1].set_xlabel(r'mitigation delay $\tau$')
axs[1].set_ylabel('seasonal amplitude $a$')
```

```python

cm = 1/2.54
fig,axs = plt.subplots(2,1,figsize=(8.6*cm,8.6*cm),height_ratios=[1,5])
axs[0].axis('off')

show, M, tauline, sline = plotting.plot_arnold(axs[1],res_a2,cmap=cmap,param='a',tol=0.015,windingnumbers=windingnumbers)

cbar = plt.colorbar(show,orientation='horizontal',ax=axs[0],location='top',aspect=15,fraction=0.8,pad=0.15)
cbar.set_ticks(np.arange(1.5,len(windingnumbers)+2))
cbar.ax.set_xticklabels(windingnumberlabels);

axs[1].set_xlabel(r'mitigation delay $\tau$')
axs[1].set_ylabel('seasonal amplitude $a$')
```
